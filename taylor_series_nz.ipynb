## Practicing Taylor Series for $\mathit{f\left(x\right) = \mathit{e}^x}$, with a (center) is not 0

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp
from scipy.special import factorial
```

```python
# Define the constants
a = -1
x = np.linspace(-6, 0, 100)
y = np.exp(x)
```

```python
# Approximate the value around 1 with second-order approximation
app = np.exp(a) + np.exp(a)*(x-a) + (np.exp(a)*(x-a)**2)/factorial(2)
plt.plot(x, y, 'k-', label=r'$\mathit{\mathit{f\left(x\right)}} = \mathit{e^x}$')
plt.plot(x, app, 'r--', label='second-order')
plt.xlim([-6.5, 6.5])
plt.legend()
plt.show()
```

```python
# Approximate the value around 1 with fourth-order approximation
app = np.exp(a) + np.exp(a)*(x-a) + (np.exp(a)*(x-a)**2)/factorial(2) + (np.exp(a)*(x-a)**3)/factorial(3) + (np.exp(a)*(x-a)**4)/factorial(4)
plt.plot(x, y, 'k-', label=r'$\mathit{\mathit{f\left(x\right)}} = \mathit{e^x}$')
plt.plot(x, app, 'r--', label='fourth-order')
plt.xlim([-6.5, 6.5])
plt.legend()
plt.show()
```
