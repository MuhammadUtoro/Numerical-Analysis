```python
## Practicing Taylor Series - cos(x) following Steve Brunton video: https://www.youtube.com/watch?v=ebfOSDj4j3I&list=PLMrJAkhIeNNTYaOnVI3QpH7jgULnAmvPA&index=6
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp
from scipy.special import factorial
```

$$\cos{x} = \mathit{1} - \frac{x^2}{2!} + \frac{x^4}{4!} - \frac{x^6}{6!} + \frac{x^8}{8!} + ... $$

```python
## Setting the exact value of sin(x)
x = np.linspace(-2*np.pi, 2*np.pi, 100)
y = np.cos(x)
```

## First-order

```python
P = [0, 0] ## Coefficient of the power from the highest, here is x
frst_order = np.polyval(P, x)
plt.plot(x, y, 'b-', label='cos(x)')
plt.plot(x, frst_order, 'k--', label='first-order')
plt.xlim([-2*np.pi, 2*np.pi])
plt.ylim([-2*np.pi, 2*np.pi])
plt.legend()
plt.grid(True)
```

## Second-order

```python
P = [-1/factorial(2), 0, 1] ## Coefficient of the power from the highest, here is x^2
scnd_order = np.polyval(P, x)
plt.plot(x, y, 'b--', label='cos(x)')
plt.plot(x, scnd_order, 'k--', label='second-order')
plt.xlim([-2*np.pi, 2*np.pi])
plt.ylim([-2*np.pi, 2*np.pi])
plt.legend()
plt.grid(True)
```

## Fourth-order

```python
P = [1/factorial(4), 0, -1/factorial(2), 0, 1] ## Coefficient of the power from the highest, here is x^4
frth_order = np.polyval(P, x)
plt.plot(x, y, 'b--', label='cos(x)')
plt.plot(x, frth_order, 'k--', label='fourth-order')
plt.xlim([-2*np.pi, 2*np.pi])
plt.ylim([-2*np.pi, 2*np.pi])
plt.legend()
plt.grid(True)
```

## Sixth-order

```python
P = [-1/factorial(6), 0, 1/factorial(4), 0, -1/factorial(2), 0, 1] ## Coefficient of the power from the highest, here is x^6
sxth_order = np.polyval(P, x)
plt.plot(x, y, 'b--', label='cos(x)')
plt.plot(x, sxth_order, 'k--', label='sixth-order')
plt.xlim([-2*np.pi, 2*np.pi])
plt.ylim([-2*np.pi, 2*np.pi])
plt.legend()
plt.grid(True)
```

## Eight-order

```python
P = [1/factorial(8) , 0, -1/factorial(6), 0, 1/factorial(4), 0, -1/factorial(2), 0, 1] ## Coefficient of the power from the highest, here is x
eigth_order = np.polyval(P, x)
plt.plot(x, y, 'b--', label='cos(x)')
plt.plot(x, eigth_order, 'k--', label='eigth-order')
plt.xlim([-2*np.pi, 2*np.pi])
plt.ylim([-2*np.pi, 2*np.pi])
plt.legend()
plt.grid(True)
```

## Plot all to compare

```python
plt.plot(x, y, 'b-', label='cos(x)')
plt.plot(x, frst_order, 'k--', label='first-order')
plt.plot(x, scnd_order, 'r--', label='second-order')
plt.plot(x, frth_order, 'm--', label='fourth-order')
plt.plot(x, sxth_order, 'g--', label='sixth-order')
plt.plot(x, eigth_order, 'y--', label='eigth-order')
plt.xlim([-2*np.pi, 2*np.pi])
plt.ylim([-2*np.pi, 2*np.pi])
plt.legend()
```
