## Practicing Taylor Series - $$\mathit{f\left(x\right) = \mathit{e^x}}$$

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp
from scipy.special import factorial
```

## Exact solution

```python
x = np.linspace(-1, 1, 100)
y = np.exp(x)
```

## First-order

```python
P = [1, 1]
frst_ord = np.polyval(P, x)
plt.plot(x, y, 'b-', label=r'$\mathit{\mathit{f\left(x\right)}} = \mathit{e^x}$')
plt.plot(x, frst_ord, 'r--', label='first-order')
plt.legend()
plt.grid(True)
plt.show()
```

## Second-order

```python
P = [1/factorial(2), 1, 1]
scnd_ord = np.polyval(P, x)
plt.plot(x, y, 'b-', label=r'$\mathit{\mathit{f\left(x\right)}} = \mathit{e^x}$')
plt.plot(x, scnd_ord, 'r--', label='second-order')
plt.legend()
plt.grid(True)
plt.show()
```

## Third-order

```python
P = [1/factorial(3), 1/factorial(2), 1, 1]
thrd_ord = np.polyval(P, x)
plt.plot(x, y, 'b-', label=r'$\mathit{\mathit{f\left(x\right)}} = \mathit{e^x}$')
plt.plot(x, thrd_ord, 'r--', label='third-order')
plt.legend()
plt.grid(True)
plt.show()
```

## fourth-order

```python
P = [1/factorial(4), 1/factorial(3), 1/factorial(2), 1, 1]
frth_ord = np.polyval(P, x)
plt.plot(x, y, 'b-', label=r'$\mathit{\mathit{f\left(x\right)}} = \mathit{e^x}$')
plt.plot(x, frth_ord, 'r--', label='fourth-order')
plt.legend()
plt.grid(True)
plt.show()
```

## fifth-order

```python
P = [1/factorial(5), 1/factorial(4), 1/factorial(3), 1/factorial(2), 1, 1]
ffth_ord = np.polyval(P, x)
plt.plot(x, y, 'b-', label=r'$\mathit{\mathit{f\left(x\right)}} = \mathit{e^x}$')
plt.plot(x, ffth_ord, 'r--', label='fifth-order')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Bonus, sixth-order
P = [1/factorial(6), 1/factorial(5), 1/factorial(4), 1/factorial(3), 1/factorial(2), 1, 1]
sxth_ord = np.polyval(P, x)
plt.plot(x, y, 'b-', label=r'$\mathit{\mathit{f\left(x\right)}} = \mathit{e^x}$')
plt.plot(x, sxth_ord, 'r--', label='sixth-order')
plt.legend()
plt.grid(True)
plt.show()
```

```python
## Plot all and compare

plt.plot(x, y, 'b-', label=r'$\mathit{\mathit{f\left(x\right)}} = \mathit{e^x}$')
plt.plot(x, frst_ord, 'r--', label='first-order')
plt.plot(x, scnd_ord, 'k--', label='second-order')
plt.plot(x, thrd_ord, 'g--', label='third-order')
plt.plot(x, frth_ord, 'm--', label='fourth-order')
plt.plot(x, ffth_ord, 'y--', label='fifth-order')
plt.plot(x, sxth_ord, 'b--', label='sixth-order')
plt.legend()
plt.grid(True)
plt.show()
```
