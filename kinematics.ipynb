Using central difference method to solve kinematics problem - consider particle moving vertically within 0,167 s time interval captured with 60 fps camera
we try to calculate the velocity at each frames and plot the position-time and velocity-time graph

```python
import numpy as np
import matplotlib.pyplot as plt
```

Input the data of the position and set up the interval and delta t

```python
t = np.zeros(11)
t_fin = 0.167
dt = np.linspace(0, t_fin, 11)
y_pos = np.array([0.0, 0.15, 0.22, 0.27, 0.30, 0.20, 0.0, -0.26, -0.30, -0.22, 0.0])
v = np.zeros(11)
a = np.zeros(11)
```

Implementing the first central different method to calculate velocity

```python
# Calculate velocities at each frame
for idx in range(1, len(y_pos) - 1):
    v[idx] = (y_pos[idx + 1] - y_pos[idx - 1]) / (dt[idx + 1] - dt[idx - 1])

# Print velocities
print("Velocities:", v)
```

Plotting Position-Time and velocity graph

```python
plt.subplot(2,1,1)
plt.plot(t+dt, y_pos, 'b-', label='Position-Time')
plt.subplot(2,1,2)
plt.plot(t+dt, v,'r-', label='Velocity-Time')
plt.show()
```

Implementing the first central different method to calculate acceleration

```python
# Calculate accelerations at each frame
for idx in range(1, len(v) - 1):
    a[idx] = (v[idx + 1] - v[idx - 1]) / (dt[idx + 1] - dt[idx - 1])

# Print velocities
print("Acceleration:", a)
```

Plotting the velocity-time and acceleration-time profile

```python
plt.subplot(2,1,1)
plt.plot(t+dt, v, 'b-', label='Velocity-Time')
plt.subplot(2,1,2)
plt.plot(t+dt, a,'r-', label='Acceleration-Time')
plt.show()
```
