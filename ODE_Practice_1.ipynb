1. $\frac{dy}{dx} = \cos{x}\sqrt{y+1}$ with the IVP $\mathit{y(\pi)=0}$

```python
#importing libraries
from scipy.integrate import odeint
from scipy.integrate import solve_ivp
import numpy as np
import matplotlib.pyplot as plt
```

```python
#define the function
def dydx1(x,y):
    return np.cos(x)*np.sqrt(y+1)

x = np.linspace(np.pi,2*np.pi,101)
y0 = 0
ysol_1 = odeint(dydx1, y0=y0, t=x, tfirst=True)

#calculate the analytical solution
yexact_1 = ((0.5*np.sin(x) + 1)**2)-1
```

```python
#plot and compare
plt.plot(x, ysol_1, 'b-', label='ode solver')
plt.plot(x, yexact_1, 'r--', label='analytical')
plt.tight_layout()
plt.xlabel('x')
plt.ylabel(r'$\frac{dy}{dx} = \cos{x}\sqrt{y+1}$')
plt.legend()
plt.show()

#we can see that the function has an analytical solution, therefore both graphs are on the top of each other
```

2. $\frac{dy}{dx} = \frac{e^y}{x^2+1}$ with IVP $\mathit{y(0)= 1}$

```python
#define the function
def dydx2(x, y):
    return np.exp(y)/(x**2+1)

x = np.linspace(0,8,101)
y0 = 1
ysol_2 = odeint(dydx2, y0=y0, t=x, tfirst=True)

#calculate the analytical solution
yexact_2 = -np.log(np.arctan(x)+np.exp(1))
```

3. $\frac{dy}{dx} = \mathcal{xy+2x+y+2}$ with IVP $\mathcal{y(2) = 0}$

```python
#define the function
def dydx3(x,y):
    return (x*y)+(2*x)+y+2

x = np.linspace(2,7,101)
y0 = 0
ysol_3 = odeint(dydx3, y0=y0, t=x, tfirst=True)

#calculate the analytical solution
yexact_3 = 2*np.exp(0.5*x**2+x-4)
```

```python
#plot and compare
plt.plot(x, ysol_3, 'b-', label='ode solver')
plt.plot(x, yexact_3, 'r--', label='analytical')
plt.tight_layout()
plt.xlabel('x')
plt.ylabel(r'$\frac{dy}{dx} = \mathcal{xy+2x+y+2}$')
plt.legend()
plt.show()

# here we can see another analytical = ODE solver
```

4. $\frac{dy}{dx} = y\log{y}$ with IVP $\mathcal{y(0) = \pi}$

```python
# Define the function
def dydx4(x, y):
    return y * np.log(y)

x = np.linspace(0,1,101)
y0 = np.pi

ysol_4 = odeint(dydx4, y0=y0, t=x, tfirst=True)

# Calculate the analytical solution
yexact_4 = np.pi**np.exp(1)**x
```

```python
# Plot and compare
plt.plot(x, ysol_4, 'b-', label='ode solver')
plt.plot(x, yexact_4, 'r--', label='analytical')
plt.tight_layout()
plt.xlabel('x')
plt.ylabel(r'$\frac{dy}{dx} = y\log{y}$')
plt.legend()
plt.show()

# Once again the numerical (ODE Solver) agrees with the analytical solution 
```

5. $\frac{dy}{dx} = \mathcal{x}\cos^2(y)+\cos^2(y)$ with IVP $\mathcal{y(0) = \frac{\pi}{4}}$

```python
# Define the function
def dydx5(x, y):
    return x * np.cos(y)**2 + np.cos(y)**2

x = np.linspace(0,1,101)
y0 = (np.pi)/4

ysol_5 = odeint(dydx5, y0=y0, t=x, tfirst=True)

# Calculate the analytical solution
yexact_5 = np.arctan((x**2/2)+x+1)
```

```python
# Plot and compare
plt.plot(x, ysol_5, 'b-', label='ode solver')
plt.plot(x, yexact_5, 'r--', label='analytical')
plt.tight_layout()
plt.xlabel('x')
plt.ylabel(r'$\frac{dy}{dx} = \mathcal{x}\cos^2(y)+\cos^2(y)$')
plt.legend()
plt.show()

# Once again the numerical (ODE Solver) agrees with the analytical solutio
```
