1. $\frac{dy}{dx} = \cos{x}\sqrt{y+1}$ with the IVP $\mathit{y(\pi)=0}$

```python
#importing libraries
from scipy.integrate import odeint
from scipy.integrate import solve_ivp
import numpy as np
import matplotlib.pyplot as plt
```

```python
#define the function
def dydx1(x,y):
    return np.cos(x)*np.sqrt(y+1)

x = np.linspace(np.pi,2*np.pi,101)
y0 = 0
ysol_1 = odeint(dydx1, y0=y0, t=x, tfirst=True)

#calculate the analytical solution
yexact_1 = ((0.5*np.sin(x) + 1)**2)-1
```

```python
#plot and compare
plt.plot(x, ysol_1, 'b-', label='ode solver')
plt.plot(x, yexact_1, 'r--', label='analytical')
plt.tight_layout()
plt.xlabel('x')
plt.ylabel(r'$\frac{dy}{dx} = \cos{x}\sqrt{y+1}$')
plt.legend()
plt.show()

#we can see that the function has an analytical solution, therefore both graphs are on the top of each other
```

3. $\frac{dy}{dx} = \mathcal{xy+2x+y+2}$ with IVP $\mathcal{y(2) = 0}$

```python
#define the function
def dydx3(x,y):
    return (x*y)+(2*x)+y+2

x = np.linspace(2,7,101)
y0 = 0
ysol_3 = odeint(dydx3, y0=y0, t=x, tfirst=True)

#calculate the analytical solution
yexact_3 = 2*np.exp(0.5*x**2+x-4)
```

```python
#plot and compare
plt.plot(x, ysol_3, 'b-', label='ode solver')
plt.plot(x, yexact_3, 'r--', label='analytical')
plt.tight_layout()
plt.xlabel('x')
plt.ylabel(r'$\frac{dy}{dx} = \mathcal{xy+2x+y+2}$')
plt.legend()
plt.show()

# here we can see another analytical = ODE solver
```

4. $\frac{dy}{dx} = y\log{y}$ with IVP $\mathcal{y(0) = \pi}$

```python
# Define the function
def dydx4(x, y):
    return y * np.log(y)

x = np.linspace(0,1,101)
y0 = np.pi

ysol_4 = odeint(dydx4, y0=y0, t=x, tfirst=True)

# Calculate the analytical solution
yexact_4 = np.pi**np.exp(1)**x
```

```python
# Plot and compare
plt.plot(x, ysol_4, 'b-', label='ode solver')
plt.plot(x, yexact_4, 'r--', label='analytical')
plt.tight_layout()
plt.xlabel('x')
plt.ylabel(r'$\frac{dy}{dx} = y\log{y}$')
plt.legend()
plt.show()

# Once again the numerical (ODE Solver) agrees with the analytical solution 
```

7. $\frac{dy}{dx} = -y + e^{x}$. This is not in the standard form, the standard form will be: $\frac{dy}{dx} + y = e^{x}$. IVP $\mathcal{y(0) = 0}$

```python
def dydx7(x, y):
    return -y + np.exp(x)

x = np.linspace(0,5,100)
y0 = 0

# Numerical solution by odeint
ysol_7 = odeint(dydx7, y0=y0, t=x, tfirst=True)

# The analytical solution
yexact_7 = (np.exp(x)/2) - 0.5*np.exp(-x)
```

```python
# Plot
plt.plot(x, ysol_7, 'b-', label='ODE solver')
plt.plot(x, yexact_7, 'r--', label='Analytical')
plt.tight_layout()
plt.xlabel('x')
plt.ylabel(r'$\frac{dy}{dx} = -y + e^{x}$')
plt.legend()
plt.show()
```

8. $\mathcal{x^2\frac{dy}{dx} + 5xy = x}$ This is not in the standard form of $\mathcal{\frac{dy}{dx} + P(x)y = q(x)}$. The standard form is: $\mathcal{\frac{dy}{dx} + \frac{5}{x}y = \frac{1}{x}}$

```python
def dydx8(x, y):
    return (-5*y)/x + (1/x)

x = np.linspace(1,6,100)
y0 = 0

# Numerical solution by odeint
ysol_8 = odeint(dydx8, y0=y0, t=x, tfirst=True)

# Analytical solution
yexact_8 = 1/5 - 1/(5*x**5)
```

```python
# Plot
plt.plot(x, ysol_8, 'b-', label='ODE solver')
plt.plot(x, yexact_8, 'r--', label='Analytical')
plt.tight_layout()
plt.xlabel('x')
plt.ylabel(r'$x^2\frac{dy}{dx} + 5xy = x$')
plt.legend()
plt.show()
```
